
# ✅ **Matplotlib 기본 차트 종류**

| 차트 종류 | 함수(메서드) | 설명 |
|-----------|-------------|------|
| 라인 차트  | `plot()` | 연속된 데이터를 라인으로 표현 |
| 산점도     | `scatter()` | 데이터를 점으로 표시하여 분포 표현 |
| 막대 차트  | `bar()` / `barh()` | 데이터의 크기를 막대로 비교 |
| 히스토그램 | `hist()` | 데이터의 빈도 및 분포 표현 |
| 파이 차트  | `pie()` | 데이터의 구성 비율 표현 |
| 박스 플롯  | `boxplot()` | 데이터 분포 및 이상치 표현 |

# ✅ **차트별 Matplotlib 예시 코드**

## ① 라인 차트 (`plot`)

| 옵션 (`loc`) | 위치                  |
|------------|----------------------|
| `'best'`   | 최적의 위치 (기본값)     |
| `'upper right'` | 우측 상단             |
| `'upper left'` | 좌측 상단             |
| `'lower right'` | 우측 하단             |
| `'lower left'` | 좌측 하단             |
| `'right'`   | 우측                 |
| `'center'`  | 정중앙                 |

---

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color="blue", linestyle="--", marker="o", label="sin(x)")
plt.title("Line Chart Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```

| 옵션                          | 설명                                 | 코드 내 예시                        |
|-------------------------------|--------------------------------------|-----------------------------------|
| **`plt.figure(figsize=(10, 6))`** | 차트의 전체 크기 설정 (가로 10, 세로 6) | `plt.figure(figsize=(10, 6))`    |
| **`plt.plot()`**              | 라인 차트를 그리는 기본 함수          | `plt.plot(x, y1)`                |
| `color`                       | 선의 색상 지정                       | `color='blue'`, `color='red'`    |
| `linestyle`                   | 선의 스타일 설정                     | 실선:`'-'`, 점선:`'--'`          |
| `marker`                      | 데이터 포인트 마커 모양 설정          | 원형:`'o'`, x형:`'x'`            |
| **`plt.title()`**             | 차트의 제목 설정                     | `plt.title('Two Line Plots: sin(x) and cos(x)')` |
| **`plt.xlabel()` / `plt.ylabel()`** | x축, y축의 라벨 설정             | `plt.xlabel('X-axis')`           |
| **`plt.xlim()` / `plt.ylim()`** | x축, y축 범위 지정                 | `plt.xlim(0,10)`                 |
| **`plt.legend(loc='upper right')`** | 범례 표시 및 위치 설정            | `plt.legend(loc='upper right')`  |
| **`plt.grid(True)`**          | 격자 표시                            | `plt.grid(True)`                 |

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 차트 그리기
plt.figure(figsize=(10, 6))
plt.plot(x, y1, color="blue", linestyle="-", marker="o", label="sin(x)")
plt.plot(x, y2, color="red", linestyle="--", marker="x", label="cos(x)")

# 옵션 설정
plt.title("Two Line Plots: sin(x) and cos(x)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)
plt.legend(loc="upper right")
plt.grid(True)

# 표시
plt.show()
```


| 옵션                          | 설명                                 | 코드 내 예시                        |
|-------------------------------|--------------------------------------|-----------------------------------|
| **`plt.figure(figsize=(10, 6))`** | 차트 전체 크기 설정 (가로 10, 세로 6)   | `plt.figure(figsize=(10, 6))`    |
| **`plt.plot()`**              | 라인 차트를 그리는 기본 함수          | `plt.plot(x, y1)`                |
| `color`                       | 선 색상 지정                         | 초록:`'green'`, 주황:`'orange'`, 보라:`'purple'` |
| `linestyle`                   | 선의 스타일 설정                     | 실선:`'-'`, 점선:`'--'`, 점-대시선:`'-.'` |
| **`plt.title()`**             | 차트 제목 설정                       | `plt.title('Three Line Plots: Exponential Functions')` |
| **`plt.xlabel()` / `plt.ylabel()`** | x축, y축 라벨 설정               | `plt.xlabel('X-axis')`           |
| **`plt.xlim()` / `plt.ylim()`** | x축, y축 범위 지정                 | `plt.xlim(0,5)`, `plt.ylim(0,150)` |
| **`plt.legend(loc='upper left')`** | 범례 표시 및 위치 설정            | `plt.legend(loc='upper left')`   |
| **`plt.grid(True)`**          | 격자 표시                            | `plt.grid(True)`                 |

---

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 5, 100)
y1 = np.exp(x)  # 지수함수
y2 = np.exp(x / 2)  # 지수함수(2로 나눈 값)
y3 = np.exp(x / 3)  # 지수함수(3으로 나눈 값)

# 차트 그리기
plt.figure(figsize=(10, 6))
plt.plot(x, y1, color="green", linestyle="-", label="exp(x)")
plt.plot(x, y2, color="orange", linestyle="--", label="exp(x/2)")
plt.plot(x, y3, color="purple", linestyle="-.", label="exp(x/3)")

# 옵션 설정
plt.title("Three Line Plots: Exponential Functions")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.xlim(0, 5)
plt.ylim(0, 150)
plt.legend(loc="upper left")
plt.grid(True)

# 표시
plt.show()
```

| 스타일      | 설명      | 예시 |
|------------|-----------|-------|
| `'-'`      | 실선 (solid)   | `linestyle='-'` |
| `'--'`     | 점선 (dashed)  | `linestyle='--'` |
| `'-.'`     | 점-대시선 (dash-dot) | `linestyle='-.'` |
| `':'`      | 짧은 점선 (dotted)  | `linestyle=':'` |


## ② 산점도 (`scatter`)

**마커(`marker`) 예시:**

| 마커       | 설명      | 예시 |
|------------|-----------|-------|
| `'o'`      | 원형 마커      | `marker='o'` |
| `'x'`      | X 모양 마커     | `marker='x'` |
| `'s'`      | 사각형 마커     | `marker='s'` |
| `'D'`      | 다이아몬드 모양 마커 | `marker='D'` |
| `'^'`      | 삼각형 마커     | `marker='^'` |

---

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color="red", alpha=0.7, marker="x")
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

## ③ 막대 차트 (`bar`)

```python
import matplotlib.pyplot as plt

# 데이터 생성
categories = ["A", "B", "C", "D"]
values = [20, 35, 30, 25]

plt.figure(figsize=(8, 6))
plt.bar(categories, values, color="green")
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

## ④ 히스토그램 (`hist`)

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
data = np.random.randn(1000)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color="purple", alpha=0.7)
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

## ⑤ 파이 차트 (`pie`)

```python
import matplotlib.pyplot as plt

# 데이터 생성
labels = ["Apple", "Banana", "Cherry", "Date"]
sizes = [15, 30, 45, 10]

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90, explode=[0.05, 0, 0, 0])
plt.title("Pie Chart Example")
plt.axis("equal")  # 원형 유지
plt.show()
```

```python
## ⑥ 박스 플롯 (`boxplot`)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=["A", "B", "C"])
plt.title("Boxplot Example")
plt.xlabel("Group")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
## ⑦ 여러개의 차트 (Subplots) 활용 예시
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 라인 차트
axes[0, 0].plot(x, y1, color="blue")
axes[0, 0].set_title("Line Chart (sin)")

# 산점도
axes[0, 1].scatter(x, y2, color="red")
axes[0, 1].set_title("Scatter Plot (cos)")

# 막대 차트
axes[1, 0].bar(["A", "B", "C"], [5, 7, 3], color="green")
axes[1, 0].set_title("Bar Chart")

# 히스토그램
axes[1, 1].hist(np.random.randn(1000), bins=20, color="purple")
axes[1, 1].set_title("Histogram")

# 공통 적용 설정 (for문 이용한 방식 예시)
for ax in axes.flat:
    ax.grid(True)

plt.tight_layout()
plt.show()
```
