### üîç **ÏöîÏïΩ Î∞è Ï∂îÏ≤ú ÌôúÏö©Î≤ï**
- ÏãúÍ∞ÅÌôîÎ•º Ìï† Îïå **`data`, `x`, `y`, `hue`** ÏòµÏÖòÏùÄ ÌïÑÏàòÏ†ÅÏúºÎ°ú Í∞ÄÏû• ÏûêÏ£º ÏÇ¨Ïö©Îê©ÎãàÎã§.
- Ïó¨Îü¨ Í∞úÏùò Ï∞®Ìä∏Í∞Ä Ìïú ÌôîÎ©¥Ïóê ÌëúÌòÑÎê† Îïå **ÏÑúÎ∏åÌîåÎ°Ø(`fig, axes = plt.subplots()`)** ÌôúÏö©ÏùÑ Í∂åÏû•Ìï©ÎãàÎã§.
- Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑù Ïãú Ïã†Î¢∞Íµ¨Í∞Ñ(**`ci`**), Ïù¥ÏÉÅÏπò(**`showfliers`**) ÏòµÏÖòÏùÑ ÏÉÅÌô©Ïóê ÎßûÍ≤å Ï†ÅÏ†àÌûà Ï°∞Ï†ïÌïòÏó¨ ÏÇ¨Ïö©ÌïòÏÑ∏Ïöî.

ÏúÑ ÌëúÏôÄ ÏòàÏãúÎ•º Ï∞∏Í≥†ÌïòÏó¨ Îã§ÏñëÌïú ÏòµÏÖòÏùÑ ÌÜµÌï¥ ÏõêÌïòÎäî ÌòïÌÉúÎ°ú Ï∞®Ìä∏Î•º ÏûêÏú†Î°≠Í≤å ÏÉùÏÑ±ÌïòÏã§ Ïàò ÏûàÏäµÎãàÎã§.

````python
import matplotlib.pyplot as plt
import numpy as np

# Í∞ÄÏÉÅ ÎçîÎØ∏ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Ï∞®Ìä∏ ÌÅ¨Í∏∞ ÏÑ§Ï†ï
plt.figure(figsize=(10, 6))

# ÎùºÏù∏ ÌîåÎ°Ø Í∑∏Î¶¨Í∏∞
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")

# Ï∞®Ìä∏ Ï†úÎ™© ÏÑ§Ï†ï
plt.title("Sin & Cos Functions")

# xÏ∂ï, yÏ∂ï Î†àÏù¥Î∏î ÏÑ§Ï†ï
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# xÏ∂ï, yÏ∂ï Î≤îÏúÑ ÏßÄÏ†ï
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

# Ï∂ï ÎààÍ∏à ÌöåÏ†Ñ Î∞è ÏÑ§Ï†ï (Ïòà: xÏ∂ï ÎààÍ∏à 45ÎèÑ ÌöåÏ†Ñ)
plt.xticks(rotation=-45)

# Î≤îÎ°Ä ÌëúÏãú Î∞è ÏúÑÏπò ÏÑ§Ï†ï (Ïò§Î•∏Ï™Ω ÏÉÅÎã®)
plt.legend(loc="upper right")

# Í∑∏Î¶¨Îìú ÌëúÏãú
plt.grid(True)

# Ï∞®Ìä∏ Î≥¥Ïó¨Ï£ºÍ∏∞
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Í∞ÄÏÉÅ ÎçîÎØ∏ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
data = pd.DataFrame(
    {
        "x": np.linspace(0, 10, 100),
        "y1": np.sin(np.linspace(0, 10, 100)),
        "y2": np.cos(np.linspace(0, 10, 100)),
        "y3": np.tan(np.linspace(0, 10, 100)),
    }
)

# ÎùºÏù∏ ÌîåÎ°Ø (ÏÑ∏ ÏÑ†)
sns.lineplot(x="x", y="y1", data=data, label="sin(x)", color="blue")
sns.lineplot(x="x", y="y2", data=data, label="cos(x)", color="red")
sns.lineplot(x="x", y="y3", data=data, label="tan(x)", color="green")
plt.title("Three Lines: sin(x), cos(x), tan(x)")
plt.legend()
plt.ylim(-10, 10)  # yÍ∞íÏùÑ Ï†úÌïúÌïòÏó¨ tan Ìï®ÏàòÏùò Í∑πÎã®Ï†ÅÏù∏ Í∞íÏù¥ ÌëúÏãúÎêòÏßÄ ÏïäÎèÑÎ°ù ÏÑ§Ï†ï
plt.show()
````

### 5. ÌûàÌä∏Îßµ (Heatmap) - ÏÉÅÍ¥Ä Í¥ÄÍ≥Ñ Î∂ÑÏÑù
- ÏÇ¨Ïö© Í≤ΩÏö∞: Ïó¨Îü¨ Î≥ÄÏàò Í∞ÑÏùò ÏÉÅÍ¥Ä Í¥ÄÍ≥ÑÎ•º ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú Î∂ÑÏÑùÌï† Îïå ÏÇ¨Ïö©Îê©ÎãàÎã§. ÏòàÎ•º Îì§Ïñ¥, Îß§Ï∂úÍ≥º ÌåÅ Í∞Ñ ÏÉÅÍ¥Ä Í¥ÄÍ≥ÑÎ•º Î∂ÑÏÑùÌïòÍ±∞ÎÇò, Ïó¨Îü¨ Î≥ÄÏàò Í∞ÑÏùò Í¥ÄÍ≥ÑÎ•º ÏãúÍ∞ÅÌôîÌïòÎäî Îç∞ Ïú†Ïö©Ìï©ÎãàÎã§.
- Ïã§Ï†Ñ ÏòàÏãú: Îß§Ï∂úÍ≥º ÌåÅ Í∞ÑÏùò ÏÉÅÍ¥Ä Í¥ÄÍ≥Ñ Î∂ÑÏÑù

````python
# ÏòàÏãú Îç∞Ïù¥ÌÑ∞ (pandas DataFrame)
data = {
    "total_bill": [23.68, 25.34, 17.92, 21.56, 30.88],
    "tip": [3.31, 4.08, 3.06, 3.71, 4.56],
}
df = pd.DataFrame(data)

# ÏÉÅÍ¥Ä Í¥ÄÍ≥Ñ Í≥ÑÏÇ∞
correlation_matrix = df.corr()

# ÌûàÌä∏ÎßµÏùÑ ÏÇ¨Ïö©Ìï¥ ÏÉÅÍ¥Ä Í¥ÄÍ≥Ñ ÏãúÍ∞ÅÌôî
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Í∞ÄÏÉÅ ÎçîÎØ∏ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
data = pd.DataFrame({"category": ["A", "B", "C", "D"], "value": [23, 45, 56, 78]})

# ÏÑúÎ∏åÌîåÎ°Ø ÏÉùÏÑ±
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Ï≤´ Î≤àÏß∏ Ï∞®Ìä∏: ÎßâÎåÄ Ï∞®Ìä∏
sns.barplot(x="category", y="value", data=data, ax=axes[0])
axes[0].set_title("Barplot of Categories")

# Îëê Î≤àÏß∏ Ï∞®Ìä∏: Î∞ïÏä§ÌîåÎ°Ø
sns.boxplot(x="category", y="value", data=data, ax=axes[1])
axes[1].set_title("Boxplot of Categories")

plt.tight_layout()
plt.show()
````
