### 🔍 **요약 및 추천 활용법**
- 시각화를 할 때 **`data`, `x`, `y`, `hue`** 옵션은 필수적으로 가장 자주 사용됩니다.
- 여러 개의 차트가 한 화면에 표현될 때 **서브플롯(`fig, axes = plt.subplots()`)** 활용을 권장합니다.
- 데이터 분석 시 신뢰구간(**`ci`**), 이상치(**`showfliers`**) 옵션을 상황에 맞게 적절히 조정하여 사용하세요.

위 표와 예시를 참고하여 다양한 옵션을 통해 원하는 형태로 차트를 자유롭게 생성하실 수 있습니다.

````python
import matplotlib.pyplot as plt
import numpy as np

# 가상 더미 데이터 생성
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 차트 크기 설정
plt.figure(figsize=(10, 6))

# 라인 플롯 그리기
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")

# 차트 제목 설정
plt.title("Sin & Cos Functions")

# x축, y축 레이블 설정
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# x축, y축 범위 지정
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

# 축 눈금 회전 및 설정 (예: x축 눈금 45도 회전)
plt.xticks(rotation=45)

# 범례 표시 및 위치 설정 (오른쪽 상단)
plt.legend(loc="upper right")

# 그리드 표시
plt.grid(True)

# 차트 보여주기
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 가상 더미 데이터 생성
data = pd.DataFrame(
    {
        "x": np.linspace(0, 10, 100),
        "y1": np.sin(np.linspace(0, 10, 100)),
        "y2": np.cos(np.linspace(0, 10, 100)),
    }
)

# 라인 플롯 (두 선)
sns.lineplot(x="x", y="y1", data=data, label="sin(x)", color="blue")
sns.lineplot(x="x", y="y2", data=data, label="cos(x)", color="red")
plt.title("Two Lines: sin(x) and cos(x)")
plt.legend()
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 가상 더미 데이터 생성
data = pd.DataFrame(
    {
        "x": np.linspace(0, 10, 100),
        "y1": np.sin(np.linspace(0, 10, 100)),
        "y2": np.cos(np.linspace(0, 10, 100)),
        "y3": np.tan(np.linspace(0, 10, 100)),
    }
)

# 라인 플롯 (세 선)
sns.lineplot(x="x", y="y1", data=data, label="sin(x)", color="blue")
sns.lineplot(x="x", y="y2", data=data, label="cos(x)", color="red")
sns.lineplot(x="x", y="y3", data=data, label="tan(x)", color="green")
plt.title("Three Lines: sin(x), cos(x), tan(x)")
plt.legend()
plt.ylim(-10, 10)  # y값을 제한하여 tan 함수의 극단적인 값이 표시되지 않도록 설정
plt.show()
````

### 5. 히트맵 (Heatmap) - 상관 관계 분석
- 사용 경우: 여러 변수 간의 상관 관계를 시각적으로 분석할 때 사용됩니다. 예를 들어, 매출과 팁 간 상관 관계를 분석하거나, 여러 변수 간의 관계를 시각화하는 데 유용합니다.
- 실전 예시: 매출과 팁 간의 상관 관계 분석

````python
# 예시 데이터 (pandas DataFrame)
data = {
    "total_bill": [23.68, 25.34, 17.92, 21.56, 30.88],
    "tip": [3.31, 4.08, 3.06, 3.71, 4.56],
}
df = pd.DataFrame(data)

# 상관 관계 계산
correlation_matrix = df.corr()

# 히트맵을 사용해 상관 관계 시각화
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 가상 더미 데이터 생성
data = pd.DataFrame({"category": ["A", "B", "C", "D"], "value": [23, 45, 56, 78]})

# 서브플롯 생성
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# 첫 번째 차트: 막대 차트
sns.barplot(x="category", y="value", data=data, ax=axes[0])
axes[0].set_title("Barplot of Categories")

# 두 번째 차트: 박스플롯
sns.boxplot(x="category", y="value", data=data, ax=axes[1])
axes[1].set_title("Boxplot of Categories")

plt.tight_layout()
plt.show()
````
